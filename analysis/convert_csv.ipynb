```python
import glob
import pandas as pd
import re
```

```python
entropy_files = glob.glob("/home/kengoaraki/Project/DA/SSDA_Annot_WSI/analysis/output/cluster_hist/*.png")
sample_num_files = glob.glob("/home/kengoaraki/Project/DA/SSDA_Annot_WSI/analysis/output/cluster_distrib/*.png")
```

#### Entropyがファイル名に書かれたヒストグラムのfile_pathからwsi, entropyのcsv作成

```python
def pickup(word: str, begin_ptr: str, end_ptr: str):
    r0 = re.search(begin_ptr, word)
    r1 = re.search(end_ptr, word)
    result = word[r0.span()[-1]:r1.span()[0]]
    # print(result)
    return result


entropy_list = []
wsi_list = []
for path in entropy_files:
    entropy = pickup(path, begin_ptr='_Entropy', end_ptr='_03')
    wsi_name = '03_' + pickup(path, begin_ptr='_03_', end_ptr='.png')
    
    entropy_list.append(entropy)
    wsi_list.append(wsi_name)

dic = {
    'wsi': wsi_list,
    'entropy': entropy_list
}
df = pd.DataFrame(dic)

df_s = df.sort_values('entropy', ascending=False).reset_index().drop('index', axis=1)
# df_s.to_csv('MF0003_cluster_entropy.csv')
print(df_s)
```

#### サンプル数がファイル名に書かれた特徴分布図のfile_pathからwsi, sample_numのcsv作成

```python
def pickup(word: str, begin_ptr: str, end_ptr: str):
    r0 = re.search(begin_ptr, word)
    r1 = re.search(end_ptr, word)
    result = word[r0.span()[-1]:r1.span()[0]]
    # print(result)
    return result

sample_num_list = []
wsi_list = []
for path in sample_num_files:

    sample_num = pickup(path, begin_ptr='_ele', end_ptr='_tsne')
    wsi_name = '03_' + pickup(path, begin_ptr='03_', end_ptr='_ele')
    print(f"{wsi_name}: {sample_num}")

    sample_num_list.append(sample_num)
    wsi_list.append(wsi_name)

dic = {
    'wsi': wsi_list,
    'sample_num': sample_num_list
}
df = pd.DataFrame(dic)

# df.to_csv('MF0003_cluster_sample_num.csv')
print(df)
```

```python
merged_df = pd.merge(df_s, df)
merged_df.sort_values('entropy', ascending=False).reset_index().drop('index', axis=1).to_csv('MF0003_cluster_entropy_num.csv')
```
